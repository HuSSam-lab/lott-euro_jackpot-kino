```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler
from keras.models import Sequential
from keras.layers import LSTM, Dense, Bidirectional, Dropout
```

```python
df = pd.read_csv('/content/keno_results_new.csv')
df.head()
```

```python
df.tail(5)
```

```python
df.shape
```

```python
df.info()
```

```python
df.drop(['king ball'], axis=1, inplace=True)
# df.drop(['20'], axis=1, inplace=True)
# df.drop(['19'], axis=1, inplace=True)
# df.drop(['18'], axis=1, inplace=True)
# df.drop(['17'], axis=1, inplace=True)
# df.drop(['16'], axis=1, inplace=True)
# df.drop(['15'], axis=1, inplace=True)
# df.drop(['14'], axis=1, inplace=True)
# df.drop(['13'], axis=1, inplace=True)
# df.drop(['12'], axis=1, inplace=True)

df.head()
```

```python
df.head()
```

```python
scaler = StandardScaler().fit(df.values)
transformed_dataset = scaler.transform(df.values)
transformed_df = pd.DataFrame(data=transformed_dataset, index=df.index)
```

```python

transformed_df.head()
```

```python
# All our games
number_of_rows = df.values.shape[0]
number_of_rows
```

```python
# Amount of games we need to take into consideration for prediction
window_length = 5
window_length
```

```python

# Balls counts
number_of_features = df.values.shape[1]
number_of_features
```

```python

X = np.empty([ number_of_rows - window_length, window_length, number_of_features], dtype=float)
y = np.empty([ number_of_rows - window_length, number_of_features], dtype=float)
for i in range(0, number_of_rows-window_length):
    X[i] = transformed_df.iloc[i : i+window_length, 0 : number_of_features]
    y[i] = transformed_df.iloc[i+window_length : i+window_length+1, 0 : number_of_features]
```

```python
X.shape
```

```python
y.shape
```

```python

model = Sequential()
```

```python
model.add(Bidirectional(LSTM(512, input_shape = (window_length, number_of_features), return_sequences = True)))
```

```python
model.add(Dropout(0.2))
```

```python
model.add(Bidirectional(LSTM(512, input_shape = (window_length, number_of_features), return_sequences = True)))
```

```python


model.add(Dropout(0.1))
```

```python
model.add(Bidirectional(LSTM(256, input_shape = (window_length, number_of_features), return_sequences = True)))
```

```python
model.add(Bidirectional(LSTM(256, input_shape = (window_length, number_of_features), return_sequences = False)))
```

```python
model.add(Dense(64))
```

```python

model.add(Dense(number_of_features))
```

```python
from tensorflow import keras
from tensorflow.keras.optimizers import Adam
model.compile(optimizer=Adam(learning_rate=0.0001), loss ='mse', metrics=['accuracy'])
```

```python
from keras.callbacks import EarlyStopping

es = EarlyStopping(monitor='loss', mode='min', verbose=1, patience=100)
```

```python
model.fit(x=X, y=y, batch_size=100, epochs=500, verbose=2,callbacks=es)
```
